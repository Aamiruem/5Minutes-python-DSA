### Handling Disconnected Graphs

```python
graph = {
    0: [1],
    1: [0],
    2: [3],
    3: [2]
}

def bfs_disconnected(graph):
    from collections import deque
    visited = set()
    for start in graph:               
        if start not in visited:
            q = deque([start])
            comp = []
            while q:
                node = q.popleft()
                if node not in visited:
                    visited.add(node)
                    comp.append(node)
                    q.extend(graph[node])
            print("Component:", comp)

bfs_disconnected(graph)
```

### Graphs with Self-Loops or Multiple Edges

```python
graph = {
    0: [0, 1, 1], 
    1: [0]
}

def dfs_handle_loops(node, visited, graph):
    if node in visited:
        return
    visited.add(node)
    print(node, end=" ")
    for nei in graph[node]:
        if nei != node: 
            dfs_handle_loops(nei, visited, graph)

dfs_handle_loops(0, set(), graph)
```

### Dealing with Infinite Cycles

```python
graph = {
    0: [1],
    1: [2],
    2: [0]
}

def safe_dfs(node, visited, graph):
    if node in visited:
        return
    visited.add(node)
    print(node, end=" ")
    for nei in graph[node]:
        safe_dfs(nei, visited, graph)

safe_dfs(0, set(), graph)
```

### Graphs with Multiple Components

```python
graph = {
    0: [1],
    1: [0],
    2: [3],
    3: [2],
    4: []   
}

def connected_components(graph):
    visited = set()
    comps = []

    def dfs(node, comp):
        visited.add(node)
        comp.append(node)
        for nei in graph[node]:
            if nei not in visited:
                dfs(nei, comp)

    for node in graph:
        if node not in visited:
            comp = []
            dfs(node, comp)
            comps.append(comp)

    return comps

print("Components:", connected_components(graph))
```

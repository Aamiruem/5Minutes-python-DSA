## 1. Node and List Class Structure

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None
```

# 2. Creating Nodes and Linking

```python
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

node1.next = node2
node2.next = node3


linked_list = SinglyLinkedList()
linked_list.head = node1
```

## 3. Traversal and Printing Nodes

```python
def print_list(head):
    current = head
    while current:
        print(current.data, end=' -> ')
        current = current.next
    print("None")

print_list(linked_list.head)
```

## 4. Insertion
### a. At Head

```python
def insert_at_head(head, data):
    new_node = Node(data)
    new_node.next = head
    return new_node

linked_list.head = insert_at_head(linked_list.head, 5)
print_list(linked_list.head)
```

### b. At Tail

```python
def insert_at_tail(head, data):
    new_node = Node(data)
    if not head:
        return new_node
    current = head
    while current.next:
        current = current.next
    current.next = new_node
    return head

linked_list.head = insert_at_tail(linked_list.head, 40)
print_list(linked_list.head)
```

### c. At Specific Position (0-based index)

```python
def insert_at_position(head, data, pos):
    if pos == 0:
        return insert_at_head(head, data)
    new_node = Node(data)
    current = head
    for _ in range(pos - 1):
        if not current:
            return head  
        current = current.next
    new_node.next = current.next
    current.next = new_node
    return head

linked_list.head = insert_at_position(linked_list.head, 25, 3)
print_list(linked_list.head)
```

## 5. Deletion
### a. From Head

```python
def delete_from_head(head):
    if not head:
        return None
    return head.next

linked_list.head = delete_from_head(linked_list.head)
print_list(linked_list.head)
```

### b. From Tail

```python
def delete_from_tail(head):
    if not head :
        return None
    current = head
    while current.next.next:
        current = current.next
    current.next = None
    return head

linked_list.head = delete_from_tail(linked_list.head)
print_list(linked_list.head)
```

### c. By Value

```python
def delete_by_value(head, value):
    if not head:
        return None
    if head.data == value:
        return head.next
    current = head
    while current.next and current.next.data != value:
        current = current.next
    if current.next:
        current.next = current.next.next
    return head

linked_list.head = delete_by_value(linked_list.head, 20)
print_list(linked_list.head)
```

## 6. Searching in a Linked List

```python
def search(head, target):
    current = head
    while current:
        if current.data == target:
            return True
        current = current.next
    return False

print("Searching for 25:", search(linked_list.head, 25))
print("Searching for 100:", search(linked_list.head, 100))
```

## 7. Time and Space Complexity of Operations
| Operation            | Time Complexity | Space Complexity |
|----------------------|------------------|------------------|
| Insertion at head    | O(1)             | O(1)             |
| Insertion at tail    | O(n)             | O(1)             |
| Insertion at pos     | O(n)             | O(1)             |
| Deletion from head   | O(1)             | O(1)             |
| Deletion from tail   | O(n)             | O(1)             |
| Deletion by value    | O(n)             | O(1)             |
| Search               | O(n)             | O(1)             |

## Singly vs Doubly Circular Linked Lists
| Feature             | Singly Circular      | Doubly Circular           |
|---------------------|----------------------|----------------------------|
| Pointers per node   | next                 | next, prev                 |
| Backward traversal  | Not possible         | Possible                   |
| Last node points to | head via `next`      | head via `next`, head via `prev` |


## 3. Node and Circular Linked List Structure

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None
```

## 4. Insertion Logic

```python
def insert(cll, data):
    new_node = Node(data)
    if cll.head is None:
        cll.head = new_node
        new_node.next = new_node
    else:
        temp = cll.head
        while temp.next != cll.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = cll.head

cll = CircularLinkedList()
insert(cll, 10)
insert(cll, 20)
insert(cll, 30)
```

## 5. Traversal and Loop Stopping Condition

```python
def print_cll(cll):
    if cll.head is None:
        print("List is empty")
        return
    temp = cll.head
    while True:
        print(temp.data, end=" -> ")
        temp = temp.next
        if temp == cll.head:
            break
    print("(back to head)")

print_cll(cll)
```

```python
def delete_node(cll, key):
    if cll.head is None:
        return

    current = cll.head
    prev = None

    while True:
        if current.data == key:
            if prev:
                prev.next = current.next
            else:
                last = cll.head
                while last.next != cll.head:
                    last = last.next
                if last == cll.head:
                    cll.head = None
                    return
                last.next = cll.head.next
                cll.head = cll.head.next
            return
        prev = current
        current = current.next
        if current == cll.head:
            break

delete_node(cll, 20)
```

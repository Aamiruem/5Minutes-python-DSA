## 1. Node Structure with Previous and Next Pointers

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
```

## 2. Creating and Linking Nodes

```python
dll = DoublyLinkedList()
node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

node1.next = node2
node2.prev = node1
node2.next = node3
node3.prev = node2

dll.head = node1
```

## 3. Traversing Forward and Backward

```python
def print_forward(head):
    current = head
    while current:
        print(current.data, end=' <-> ')
        last = current
        current = current.next
    print("None")
    return last 

def print_backward(tail):
    current = tail
    while current:
        print(current.data, end=' <-> ')
        current = current.prev
    print("None")


tail = print_forward(dll.head)
print_backward(tail)
```

## 4. Insertion
### a. At Head

```python
def insert_at_head(dll, data):
    new_node = Node(data)
    new_node.next = dll.head
    if dll.head:
        dll.head.prev = new_node
    dll.head = new_node

insert_at_head(dll, 5)
print_forward(dll.head)
```

### b. At Tail

```python
def insert_at_tail(dll, data):
    new_node = Node(data)
    if not dll.head:
        dll.head = new_node
        return
    current = dll.head
    while current.next:
        current = current.next
    current.next = new_node
    new_node.prev = current

insert_at_tail(dll, 40)
print_forward(dll.head)
```

### c. At Specific Index

```python
def insert_at_index(dll, data, index):
    if index == 0:
        insert_at_head(dll, data)
        return
    new_node = Node(data)
    current = dll.head
    for _ in range(index - 1):
        if not current:
            return 
        current = current.next
    if not current:
        return
    new_node.next = current.next
    if current.next:
        current.next.prev = new_node
    new_node.prev = current
    current.next = new_node

insert_at_index(dll, 25, 3)
print_forward(dll.head)
```

## 5. Deletion
### a. From Head

```python
def delete_from_head(dll):
    if not dll.head:
        return
    dll.head = dll.head.next
    if dll.head:
        dll.head.prev = None

delete_from_head(dll)
print_forward(dll.head)
```

### b. From Tail

```python
def delete_from_tail(dll):
    if not dll.head:
        return
    current = dll.head
    if not current.next:
        dll.head = None
        return
    while current.next:
        current = current.next
    current.prev.next = None

delete_from_tail(dll)
print_forward(dll.head)
```

### c. From Specific Index

```python
def delete_from_index(dll, index):
    if index == 0:
        delete_from_head(dll)
        return
    current = dll.head
    for _ in range(index):
        if not current:
            return 
        current = current.next
    if not current:
        return
    if current.prev:
        current.prev.next = current.next
    if current.next:
        current.next.prev = current.prev

delete_from_index(dll, 2)
print_forward(dll.head)
```

## 6. Comparison with Singly Linked List
| Feature                  | Singly Linked List | Doubly Linked List |
|--------------------------|--------------------|--------------------|
| Memory usage             | Less               | More (extra prev)  |
| Insertion/deletion       | Forward only       | Forward & backward |
| Traversal                | One direction      | Both directions    |
| Ease of reversal         | Difficult          | Easier             |

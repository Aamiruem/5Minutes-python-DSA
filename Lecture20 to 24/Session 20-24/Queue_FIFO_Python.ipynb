## 1. Queue Implementation
### a. Using List

```python
queue = []
queue.append(1)
queue.append(2)
print("Dequeued:", queue.pop(0))
print("Front element:", queue[0] if queue else None)
```

### b. Using collections.deque (Efficient)

```python
from collections import deque

queue = deque()
queue.append(1)
queue.append(2)
print("Dequeued:", queue.popleft())
print("Front element:", queue[0] if queue else None)
```

### c. Using Custom Class

```python
class Queue:
    def __init__(self):
        self.container = deque()

    def enqueue(self, item):
        self.container.append(item)

    def dequeue(self):
        return self.container.popleft() if not self.is_empty() else None

    def front(self):
        return self.container[0] if not self.is_empty() else None

    def is_empty(self):
        return len(self.container) == 0

    def size(self):
        return len(self.container)

q = Queue()
q.enqueue(10)
q.enqueue(20)
print("Front:", q.front())
print("Dequeued:", q.dequeue())
print("Is empty?", q.is_empty())
```

## 2. Queue Use-Cases
### a. Scheduling / Buffering Example

```python
jobs = deque(["job1", "job2", "job3"])

while jobs:
    job = jobs.popleft()
    print("Processing:", job)
```

### b. Level-Order Traversal (Binary Tree)

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None

def level_order(root):
    if not root:
        return
    q = deque([root])
    print(q)
    while q:
        node = q.popleft()
        print(node.val, end=" ")
        if node.left: q.append(node.left)
        if node.right: q.append(node.right)

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
level_order(root)
```

### b. Circular Queue

```python
class CircularQueue:
    def __init__(self, k):
        self.queue = [None] * k
        self.head = self.tail = -1
        self.size = k

    def enqueue(self, value):
        if (self.tail + 1) % self.size == self.head:
            print("Queue is full")
            return
        if self.head == -1:
            self.head = 0
        self.tail = (self.tail + 1) % self.size
        self.queue[self.tail] = value

    def dequeue(self):
        if self.head == -1:
            print("Queue is empty")
            return
        removed = self.queue[self.head]
        if self.head == self.tail:
            self.head = self.tail = -1
        else:
            self.head = (self.head + 1) % self.size
        return removed

cq = CircularQueue(3)
cq.enqueue(1)
cq.enqueue(2)
cq.enqueue(3)
cq.enqueue(4)
print("Dequeued:", cq.dequeue())
```

### c. Priority Queue (using heapq)

```python
import heapq

pq = []
heapq.heappush(pq, (3, "Low priority"))
heapq.heappush(pq, (1, "High priority"))
heapq.heappush(pq, (2, "Medium priority"))

while pq:
    print("Processing:", heapq.heappop(pq)[1])
```

### d. Double-Ended Queue (Deque)

```python
dq = deque()
dq.append(1)    
dq.appendleft(0) 
print("Dequeued right:", dq.pop())
print("Dequeued left:", dq.popleft())
```

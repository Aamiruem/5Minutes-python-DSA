## 1. Traversing Arrays

```python
arr = [10, 20, 30, 40, 50]

for x in arr:
    print(x, end=' ')
print()

i = 0
while i < len(arr):
    print(arr[i], end=' ')
    i += 1
```

## 2. Searching
### a. Linear Search

```python
arr = [5, 3, 7, 1, 9]
target = 8

found = False
for i in range(len(arr)):
    if arr[i] == target:
        print(f"Element found at index {i}")
        found = True
        break

if not found:
    print("Element not found")
```

### b. Binary Search

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [1, 3, 5, 7, 9, 11]
target = 7
index = binary_search(arr, target)
print(f"Element found at index {index}" if index != -1 else "Element not found")
```

## 3. Insertion and Deletion at Specific Positions
### a. Insertion

```python
arr = [1, 2, 4, 5]
arr.insert(2, 3) 
print("After insertion:", arr)
```

### b. Deletion

```python
arr = [1, 2, 3, 4, 5]
del arr[2] 
print("After deletion:", arr)
```

## 4. Reversing an Array

```python
arr = [1, 2, 3, 4, 5]
arr.reverse()
print("Reversed array:", arr)
```

## 5. Rotating Arrays
### a. Left Rotation

```python
arr = [1, 2, 3, 4, 5]
k = 1
k %= len(arr)
arr = arr[k:] + arr[:k]
print("Left rotated array:", arr)
```

### b. Right Rotation

```python
arr = [1, 2, 3, 4, 5]
k = 1
k %= len(arr)
arr = arr[-k:] + arr[:-k]
print("Right rotated array:", arr)
```

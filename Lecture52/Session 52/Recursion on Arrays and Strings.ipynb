### Linear recursion on arrays

```python
def print_arr(arr,i=0):
    if i==len(arr): return
    print(arr[i])
    print_arr(arr,i+1)

print_arr([1,2,3,4])
```

### Recursive search in an array

```python
def search(arr,x,i=0):
    if i==len(arr): return False
    if arr[i]==x: return True
    return search(arr,x,i+1)

print(search([1,2,3,4],3)) 
print(search([1,2,3,4],5))  
```

### Finding max/min in array recursively

```python
def arr_max(arr,i=0):
    if i==len(arr)-1: return arr[i]
    return max(arr[i], arr_max(arr,i+1))

def arr_min(arr,i=0):
    if i==len(arr)-1: return arr[i]
    return min(arr[i], arr_min(arr,i+1))

a=[5,1,8,3,2]
print(arr_max(a)) 
print(arr_min(a))  
```

### Reversing arrays using recursion

```python
def reverse(arr,l=0,r=None):
    if r is None: r=len(arr)-1
    if l>=r: return
    arr[l],arr[r]=arr[r],arr[l]
    reverse(arr,l+1,r-1)

a=[1,2,3,4,5]
reverse(a)
print(a) 
```

### Recursion on strings (palindrome check example)

```python
def is_pal(s,l=0,r=None):
    if r is None: r=len(s)-1
    if l>=r: return True
    if s[l]!=s[r]: return False
    return is_pal(s,l+1,r-1)

print(is_pal("madam"))  
print(is_pal("hello"))  
```

### Heap implementation using an array

```python
def sift_up(a, i):
    while i > 0:
        p = (i - 1) // 2
        if a[p] <= a[i]:
            break
        a[p], a[i] = a[i], a[p]
        i = p


def sift_down(a, i):
    n = len(a)
    while True:
        l, r = 2 * i + 1, 2 * i + 2
        smallest = i
        if l < n and a[l] < a[smallest]:
            smallest = l
        if r < n and a[r] < a[smallest]:
            smallest = r
        if smallest == i:
            break
        a[i], a[smallest] = a[smallest], a[i]
        i = smallest
```

```python
def build_min_heap(a):
    for i in range((len(a)//2) - 1, -1, -1):
        sift_down(a, i)
```

```python
a = [7, 3, 9, 1, 4]
build_min_heap(a) 
```

### Insertion and  Deletion

```python
def heappush(a, x):
    a.append(x)
    sift_up(a, len(a) - 1)

def heappop(a):
    a[0], a[-1] = a[-1], a[0]
    x = a.pop()
    if a:
        sift_down(a, 0)
    return x
```

```python
h = []
heappush(h, 5)
heappush(h, 2)
heappush(h, 8)
print(h)          
print(heappop(h)) 
print(heappop(h)) 
```

### Python’s heapq (min-heap by default)

```python
import heapq

h = []
heapq.heappush(h, 5)
heapq.heappush(h, 2)
heapq.heappush(h, 8)
print(heapq.heappop(h))  
```

```python
a = [7, 3, 9, 1, 4]
heapq.heapify(a)
print(a[0])      
print(heapq.heappop(a)) 
```

```python
import heapq

h = []
for x in [5, 2, 8]:
    heapq.heappush(h, -x)      

max_val = -heapq.heappop(h)    
print(max_val)               
```

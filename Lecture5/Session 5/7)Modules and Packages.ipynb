## 1. Importing Built‑in Modules

````python
import math, random, datetime
print('π ≈', math.pi)
print('Random 1‑6 =>', random.randint(1, 6))
print('Now ⇒', datetime.datetime.now())
````

## 2. Creating & Importing Custom Modules

Just save a `.py` file and import it!

````python
%%writefile my_utils.py

def greeting(name):
    return f"Hello, {name}!"
````

````python
a=1
f"Hello, {a}!"
````

````python
a=1
f"Hello, {a+1},{a*2},{a/1}!"
````

````python
a=1
f"Hello, {1}(1)!"
````

````python
a=1
f"Hello, {1}[1]!"
````

````python
import importlib, my_utils
importlib.reload(my_utils)
print(my_utils.greeting('Shridhar'))
````

## 5. Python Package Index (PyPI) & `pip`

- **PyPI** is the central repository for third‑party Python packages.  
- Use **`pip`** to install packages from PyPI.

```bash
# install the popular 'numpy'
pip install numpy

# upgrade pip itself
python -m pip install --upgrade pip
```

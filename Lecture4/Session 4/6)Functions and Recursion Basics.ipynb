# Python Functions & Recursion

## 1. Defining and Calling Functions

```python
def greet(name):
    print("Hello", name)

greet("Shridhar")
```

## 2. Parameters and Return Values

```python
def square(x):
    return x * x

print('4 squared is', square(4))
```

## 3. Default, Keyword, and Variable-Length Arguments

### Default Arguments

```python
def greet(name, message='Hi'):
    print(f"{message}, {name}!")

greet('Shridhar')
```

### Keyword Arguments

```python
greet(message='Hello', name='Shridhar')
```

### Variable-Length Arguments

```python
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))
```

```python
def print_details(**kwargs):
    for k, v in kwargs.items():
        print(k, ':', v)

print_details(name='Shridhar', age=25)
```

## 4. Scope: Local vs Global Variables

```python
count = 0

def increment():
    count1 = 5
    count1 += 1
    print(count1)

increment()
print('Count:', count)
```

```python
count = 0

def increment():
    count = 5
    count += 1
    print(count)

increment()
print('Count:', count)
```

```python
count = 0

def increment():
    global count
    count += 1

increment()
print('Count:', count)
```

## 5. Recursion Fundamentals and Base Case

```python
def countdown(n):
    if n == 0:
        print('Gooo!')
    else:
        print(n)
        countdown(n - 1)

countdown(3)
```

## 6. Simple Recursion Examples

### Factorial

```python
def factorial(n):
    print(n)
    return 1 if n == 0 else n * factorial(n - 1)
factorial(3)
```

### Fibonacci (nth number)

```python
def fibonacci(n):
    return n if n <= 1 else fibonacci(n - 1) + fibonacci(n - 2)
fibonacci(3)
```

### Sum of Digits

```python
def sum_digits(n):
    print(n)
    return 0 if n == 0 else n % 10 + sum_digits(n // 10)
sum_digits(12)
```

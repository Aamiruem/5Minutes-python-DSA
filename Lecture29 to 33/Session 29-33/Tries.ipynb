```python
class TrieNode:
    def __init__(self):
        self.children = {}      
        self.is_end = False     
```

```python
class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        node = self.root
        for ch in word:
            if ch not in node.children:
                node.children[ch] = TrieNode()
            node = node.children[ch]
        node.is_end = True
```

```python
class Trie(Trie):
    def search(self, word: str) -> bool:
        node = self.root
        for ch in word:
            if ch not in node.children:
                return False
            node = node.children[ch]
        return node.is_end 

    def starts_with(self, prefix: str) -> bool:
        node = self.root
        for ch in prefix:
            if ch not in node.children:
                return False
            node = node.children[ch]
        return True  
```

```python
class Trie(Trie):
    def delete(self, word: str) -> None:
        def _delete(node: TrieNode, i: int) -> bool:
            if i == len(word):
                if not node.is_end:       
                    return False        
                node.is_end = False      
                return len(node.children) == 0  

            ch = word[i]
            if ch not in node.children:
                return False             

            should_prune_child = _delete(node.children[ch], i + 1)

            if should_prune_child:
                del node.children[ch]

            return (not node.is_end) and (len(node.children) == 0)

        _ = _delete(self.root, 0)
```

```python
trie = Trie()

# Insert
for w in ["cat", "car", "care", "dog"]:
    trie.insert(w)

# Search
print(trie.search("car"))   
print(trie.search("cap"))  

# Prefix
print(trie.starts_with("ca"))  
print(trie.starts_with("do"))  
print(trie.starts_with("z"))  

# Delete
trie.delete("car")
print(trie.search("car"))  
print(trie.search("care")) 
print(trie.search("cat"))   
```

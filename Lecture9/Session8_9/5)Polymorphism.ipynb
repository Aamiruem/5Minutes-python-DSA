## 1. Method Overriding
* Child class provides a different implementation for a method from the parent class.

```python
class Vehicle:
    def start(self):
        return "Vehicle starting"

class Car(Vehicle):
    def start(self):
        return "Car starting with ignition"

v = Vehicle()
c = Car()
print(v.start())
print(c.start())
```

## 2. Operator Overloading (`__add__`, `__eq__`, etc.)
* Python lets you redefine built-in operators for your custom classes.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
p3 = p1 + p2
print(p3)
print(p3 == Point(4, 6))
```

## 3. Duck Typing in Python
* "If it looks like a duck and quacks like a duck, it must be a duck."
* Focuses on behavior rather than explicit inheritance.

```python
class Duck:
    def quack(self):
        print("Quack!")

class Person:
    def quack(self):
        print("I'm quacking like a duck!")

def make_it_quack(thing):
    thing.quack()

make_it_quack(Duck())
make_it_quack(Person())
```

## 4. Function Overloading

```python
def greet(name):
    print(f"Hello, {name}")

def greet():
    print("Hello!")

greet()
```

```python
def greet(name=None):
    if name:
        print(f"Hello, {name}")
    else:
        print("Hello!")

greet()            
greet("Shridhar")
```

```python
def greet(*args):
    if len(args) == 0:
        print("Hello!")
    elif len(args) == 1:
        print(f"Hello, {args[0]}!")
    else:
        print("Hello everyone:", ', '.join(args))

greet()
greet("Shridhar")
greet("Asha", "John", "Meera")
```

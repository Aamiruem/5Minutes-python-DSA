## Problem 1 – Grade Calculator  
Write a script that asks the user for their exam score (0 – 100) and prints the corresponding grade:
- 90 – 100 → **A**  
- 80 – 89 → **B**  
- 70 – 79 → **C**  
- Below 70 → **D**  

Use `if`, `elif`, and `else`
Handle invalid (non‑numeric or out‑of‑range) input with a friendly message.

```python
def grade_from_score(score: int):
    """Return a letter grade for the given numeric score."""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

try:
    raw = input('Enter marks (0‑100): ')
    score = int(raw)
    if not 0 <= score <= 100:
        raise ValueError('Score must be between 0 and 100')
    print('Your grade →', grade_from_score(score))
except ValueError as err:
    print('Invalid input:', err)
```

## Problem 2 – Custom Math Utility Module  

1. Create a module **`math_utils.py`** implementing:
   - `factorial(n)` (recursive)
   - `is_prime(n)`  
2. Import and demonstrate both functions.
3. Handle negative input for `factorial` using `raise`.

```python
%%writefile math_utils.py
def factorial(n):
  try:
    if n < 0:
        raise ValueError('Negative numbers have no factorial')
    return 1 if n == 0 else n * factorial(n - 1)
    
  except ValueError as err:
    print('Invalid input:', err)

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

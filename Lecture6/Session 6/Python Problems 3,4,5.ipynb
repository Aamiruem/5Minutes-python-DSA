# Problem 3: Count Vowels and Consonants
1. Concepts: `loops`, `if-else`                                                
2. Problem: Take a string input and count the number of vowels and consonants.

```python
text = input("Enter a string: ").lower()
vowels = "aeiou"
vowel_count = consonant_count = 0

for char in text:
    if char.isalpha():
        if char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

# Problem 4: Sum of Digits
1. Concepts: `while` loop
2. Problem: Take an integer and find the sum of its digits.

```python
num = int(input("Enter a number: "))
total = 0
while num > 0:
    total += num % 10
    num //= 10
print("Sum of digits:", total)
```

# Problem 5: Reverse a String using loop
1. Concepts: `loops`
2. Problem: Input a string. Print its reverse using a loop.

```python
text = input("Enter a string: ")
reversed_text = ""

for char in text:
    reversed_text = char + reversed_text

print("Reversed string:", reversed_text)
```

## 1. Strings in Python are Immutable

```python
s = "hello"
try:
    s[0] = 'H'  
except TypeError as e:
    print("Error:", e)

s = 'H' + s[1:]
print("Modified string:", s)
```

## 2. Creating and Accessing Strings

```python
s = "Python Programming"
print("First character:", s[0])
print("Last character:", s[-1])
print("Length of string:", len(s))
```

## 3. Slicing and Indexing

```python
s = "Hello, World!"
print("s[0:5]:", s[0:5])  
print("s[:5]:", s[:5])    
print("s[7:]:", s[7:])    
print("s[::-1]:", s[::-1])  
```

## 4. String Methods

```python
s = "  Hello, Python!  "
print("Original:", repr(s))
print("lower():", s.lower())
print("upper():", s.upper())
print("strip():", s.strip())

words = s.strip().split()
print("split():", words)

joined = "-".join(words)
print("join():", joined)

replaced = s.replace("Python", "World")
print("replace():", replaced)
```

## 5. String Formatting
### a. f-strings (Python 3.6+ syntax)

```python
name = "shridhar"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

### b. .format() Method

```python
name = "shridhar"
language = "Python"
print("Hello, my name is {} and I code in {}.".format(name, language))
```

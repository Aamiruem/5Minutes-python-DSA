## 1. Reversing Words and Characters

```python
s = "hello world"

rev_str = s[::-1]
print("Reversed string:", rev_str)

words = s.split()
reversed_words = [word[::-1] for word in words]
print("Reversed words:", " ".join(reversed_words))
```

## 2. Checking for Palindrome

```python
s = "madam"
is_palindrome = s == s[::-1]
print(f"'{s}' is a palindrome:", is_palindrome)
```

## 3. Checking for Anagram

```python
from collections import Counter

s1 = "listen"
s2 = "silent"
is_anagram = Counter(s1) == Counter(s2)
print(f"'{s1}' and '{s2}' are anagrams:", is_anagram)
```

## 4. Checking for Substring

```python
s = "hello world"
substr = "world"
is_substring = substr in s
print(f"'{substr}' is in '{s}':", is_substring)
```

## 5. Character Frequency Count using Dictionary

```python
s = "programming"
freq = {}

for char in s:
    if char in freq:
      freq[char] += 1
    else:
      freq[char] = 1

print("Character frequencies:")
for char, count in freq.items():
    print(f"{char}: {count}")
```

## 6. Removing Duplicates from a String

```python
s = "programming"
result = ""
seen = set()

for char in s:
    if char not in seen:
        seen.add(char)
        result += char

print("String after removing duplicates:", result)
```

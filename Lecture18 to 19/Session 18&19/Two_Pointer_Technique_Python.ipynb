## Finding Pairs with a Given Sum

```python
arr = [1, 2, 3, 4, 6, 8, 9]
target = 10
left, right = 0, len(arr) - 1

while left < right:
    curr_sum = arr[left] + arr[right]
    if curr_sum == target:
        print(f"Pair found: ({arr[left]}, {arr[right]})")
        break
    elif curr_sum < target:
        left += 1
    else:
        right -= 1
else:
    print("No pair found")
```

## Removing Duplicates from a Sorted Array

```python
arr = [1, 1, 2, 2, 3, 4, 4, 5]
if not arr:
    print("Empty list")
else:
    i = 0
    for j in range(1, len(arr)):
        if arr[j] != arr[i]:
            i += 1
            arr[i] = arr[j]
    result = arr[:i+1]
    print("After removing duplicates:", result)
```

## Even-Odd Partition

```python
arr = [12, 17, 70, 15, 22, 65, 21, 90]
left, right = 0, len(arr) - 1

while left < right:
    while arr[left] % 2 == 0 and left < right:
        left += 1
    while arr[right] % 2 == 1 and left < right:
        right -= 1
    if left < right:
        arr[left], arr[right] = arr[right], arr[left]

print("Even-Odd partitioned array:", arr)
```

## Positive-Negative Partition

```python
arr = [12, -7, 5, -3, 9, -2, -1, 6]
left, right = 0, len(arr) - 1

while left < right:
    while arr[left] < 0 and left < right:
        left += 1
    while arr[right] >= 0 and left < right:
        right -= 1
    if left < right:
        arr[left], arr[right] = arr[right], arr[left]

print("Positive-Negative partitioned array:", arr)
```

## Merging Two Sorted Arrays In-place

```python
arr1 = [1, 3, 5, 0, 0, 0]
arr2 = [2, 4, 6]
m, n = 3, 3

i, j, k = m - 1, n - 1, m + n - 1
while i >= 0 and j >= 0:
    if arr1[i] > arr2[j]:
        arr1[k] = arr1[i]
        i -= 1
    else:
        arr1[k] = arr2[j]
        j -= 1
    k -= 1

while j >= 0:
    arr1[k] = arr2[j]
    j -= 1
    k -= 1

print("Merged array:", arr1)
```

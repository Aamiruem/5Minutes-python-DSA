## 1. Fixed-size Sliding Window
### a. Maximum Sum Subarray of Size K

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

max_sum = curr_sum = sum(arr[:k])
for i in range(k, len(arr)):
    curr_sum += arr[i] - arr[i - k]
    max_sum = max(max_sum, curr_sum)

print("Maximum sum of subarray of size", k, "is", max_sum)
```

### b. Minimum Sum Subarray of Size K

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

min_sum = curr_sum = sum(arr[:k])
for i in range(k, len(arr)):
    curr_sum += arr[i] - arr[i - k]
    min_sum = min(min_sum, curr_sum)

print("Minimum sum of subarray of size", k, "is", min_sum)
```

## 2. Variable-size Sliding Window
### a. Longest Substring with Unique Characters

```python
s = "abcabcbb"
char_index = {}
left = 0
max_len = 0

for right in range(len(s)):
    if s[right] in char_index and char_index[s[right]] >= left:
        left = char_index[s[right]] + 1
    char_index[s[right]] = right
    max_len = max(max_len, right - left + 1)

print("Length of longest substring with unique characters:", max_len)
```

### b. Longest Subarray with Sum ≤ K

```python
arr = [1, 2, 1, 0, 1, 1, 0]
k = 4
left = 0
curr_sum = 0
max_len = 0

for right in range(len(arr)):
    curr_sum += arr[right]
    while curr_sum > k:
        curr_sum -= arr[left]
        left += 1
    max_len = max(max_len, right - left + 1)

print("Longest subarray with sum ≤", k, "is of length", max_len)
```

## 3. Optimizing Brute Force using Sliding Window

```python
arr = [2, 1, 5, 1, 3, 2]
k = 3

max_sum_bf = float('-inf')
for i in range(len(arr) - k + 1):
    curr_sum = sum(arr[i:i+k])
    max_sum_bf = max(max_sum_bf, curr_sum)

print("Brute-force max sum:", max_sum_bf)

curr_sum = sum(arr[:k])
max_sum_sw = curr_sum
for i in range(k, len(arr)):
    curr_sum += arr[i] - arr[i-k]
    max_sum_sw = max(max_sum_sw, curr_sum)

print("Sliding Window max sum:", max_sum_sw)
```

## 4. Pointer to Pointer (`**ptr`) Simulation

```python
class Box:
    def __init__(self, value):
        self.value = value

x = Box(100)
ptr = x
ptr_to_ptr = [ptr]

print("Original value:", ptr_to_ptr[0].value)

ptr_to_ptr[0].value = 200
print("Modified via **ptr simulation:", x.value)
```

## 1. Call by Value vs Call by Reference

```python
def modify_val(x):
    x = 100  
    print("Inside modify_val:", x)

def modify_list(lst):
    lst[0] = 100 

a = 10
modify_val(a)
print("After modify_val:", a)

lst = [10, 20, 30]
modify_list(lst)
print("After modify_list:", lst)
```

## 2. Returning Multiple Values via 'Pointers' (Tuples)

```python
def calculate(x, y):
    sum_val = x + y
    diff = x - y
    return sum_val, diff  

a, b = 10, 5
s, d = calculate(a, b)
print("Sum:", s, "Difference:", d)
```

## 3. Swapping Using Pointers (References in List)

```python
def swap(values):
    values[0], values[1] = values[1], values[0]

pair = [10, 20]
swap(pair)
print("Swapped values:", pair)
```

## 4. Function Pointers (Storing Functions in Variables)

```python
def greet():
    print("Hello!")

def farewell():
    print("Goodbye!")

ptr = greet
ptr()
ptr = farewell
ptr()
```

## 5. Passing Functions as Parameters

```python
def execute(func):
    print("Calling passed function:")
    func()

execute(greet)
execute(farewell)
```

# Slow and Fast Pointer Techniques

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def print_list(head):
    while head:
        print(head.data, end=" -> ")
        head = head.next
    print("None")
```

## 1. Cycle Detection (Floyd’s Tortoise and Hare)

```python
def has_cycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = head.next 

print("Cycle detected?" , has_cycle(head))
```

## 2. Find Middle of Linked List

```python
def find_middle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow


head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

mid = find_middle(head)
print("Middle node:", mid.data)
```

## 3. Detect Intersection in Two Linked Lists

```python
def get_length(head):
    length = 0
    while head:
        length += 1
        head = head.next
    return length

def get_intersection_node(headA, headB):
    lenA, lenB = get_length(headA), get_length(headB)
    while lenA > lenB:
        headA = headA.next
        lenA -= 1
    while lenB > lenA:
        headB = headB.next
        lenB -= 1
    while headA and headB:
        if headA == headB:
            return headA
        headA = headA.next
        headB = headB.next
    return None


intersect = Node(8)
intersect.next = Node(10)

a = Node(3)
a.next = Node(7)
a.next.next = intersect

b = Node(99)
b.next = intersect

node = get_intersection_node(a, b)
print("Intersection at node:", node.data if node else "No intersection")
```

## 4. Remove N-th Node from End

```python
def remove_nth_from_end(head, n):
    dummy = Node(0)
    dummy.next = head
    slow = fast = dummy

    for _ in range(n):
        fast = fast.next

    while fast.next:
        slow = slow.next
        fast = fast.next

    slow.next = slow.next.next
    return dummy.next


head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

new_head = remove_nth_from_end(head, 2)
print("After removing 2nd node from end:")
print_list(new_head)
```

## 1) Structure of a Binary Tree Node (left, right, data)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None  
        self.right = None  

    def __repr__(self):
        return f"Node({self.data})"
```

## 2) Creating a Binary Tree

```python
# Example tree:
#        1
#      /   \
#     2     3
#    / \     \
#   4   5     6

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.right = Node(6)

print("Root:", root)
print("Root.left:", root.left)
print("Root.right:", root.right)
```

## 3) Tree Traversal

```python
def inorder(root):
    res = []
    def dfs(node):
        if not node: return
        dfs(node.left)
        res.append(node.data)
        dfs(node.right)
    dfs(root)
    return res

print("Inorder:", inorder(root))  
```

```python
def preorder(root):
    res = []
    def dfs(node):
        if not node: return
        res.append(node.data)
        dfs(node.left)
        dfs(node.right)
    dfs(root)
    return res

print("Preorder:", preorder(root))
```

```python
def postorder(root):
    res = []
    def dfs(node):
        if not node: return
        dfs(node.left)
        dfs(node.right)
        res.append(node.data)
    dfs(root)
    return res

print("Postorder:", postorder(root))
```

```python
from collections import deque

def level_order(root):
    if not root:
        return []
    q = deque([root])
    levels = []
    while q:
        sz = len(q)
        level = []
        for _ in range(sz):
            node = q.popleft()
            level.append(node.data)
            if node.left: q.append(node.left)
            if node.right: q.append(node.right)
        levels.append(level)
    return levels

print("Level order (by levels):", level_order(root))
```

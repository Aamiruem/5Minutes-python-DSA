## 1. itertools.count()

```python
import itertools

counter = itertools.count(10, 2)
for num in counter:
    if num > 20: 
        break
    print(num)
```

## 2. itertools.cycle()

```python
import itertools

cycled = itertools.cycle([1, 2, 3])
for _ in range(7): 
    print(next(cycled))
```

## 3. itertools.repeat()

```python
import itertools

repeated = itertools.repeat('A', 5)
print(repeated)
for char in repeated:
    print(char)
```

## 4. itertools.combinations()

```python
import itertools

data = [1, 2, 3, 4]
comb = itertools.combinations(data, 2)
print(list(comb))
```

## 5. itertools.permutations()

```python
import itertools

data = [1, 2, 3]
perm = itertools.permutations(data, 2)
print(list(perm))
```

## 6. itertools.product()

```python
import itertools

data1 = [1, 2]
data2 = ['A', 'B']
prod = itertools.product(data1, data2)
print(list(prod))
```

## 7. itertools.chain()

```python
import itertools

data1 = [1, 2, 3]
data2 = ['A', 'B']
chained = itertools.chain(data1, data2)
print(chained)
print(list(chained))
```

## 8. itertools.islice()

```python
import itertools

data = [1, 2, 3, 4, 5]
sliced = itertools.islice(data, 0, 8)
print(list(sliced))
```

## 9. itertools.groupby()

```python
import itertools

data = [1, 1, 2, 2, 3, 3, 3]
grouped = itertools.groupby(data)
for key, group in grouped:
    print(key, list(group))
```

## 10. itertools.compress()

```python
import itertools

data = ['A', 'B', 'C', 'D', 'E']
selectors = [1, 0, 1, 0, 1]  
result = itertools.compress(data, selectors)
print(list(result))
```

## 12. itertools.tee()

```python
import itertools

data = [1, 2, 3, 4, 5]
it1, it2 = itertools.tee(data, 2)

print(list(it1)) 
print(list(it2))
```

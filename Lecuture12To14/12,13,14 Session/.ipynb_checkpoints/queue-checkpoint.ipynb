## queue.Queue()

```python
import queue

q = queue.Queue()
q.put(1)
q.put(2)
q.put(3)
print(q.get())
```

## queue.LifoQueue()

```python
q = queue.LifoQueue()
q.put(1)
q.put(2)
q.put(3)
print(q.get())  
```

## queue.PriorityQueue()

```python
q = queue.PriorityQueue()
q.put((1, 'task1'))
q.put((3, 'task3'))
q.put((2, 'task2'))
print(q.get()) 
```

## put()

```python
q = queue.Queue()
q.put(10)
print(q.queue)
q.put(20)
print(q.queue)
```

## get()

```python
q = queue.Queue()
q.put(10)
item = q.get()
print(item)  
```

## empty()

```python
q = queue.Queue()
q.put(10)
print(q.empty())  
q.get()
print(q.empty()) 
```

## full()

```python
q = queue.Queue(maxsize=2)
q.put(1)
q.put(2)
print(q.full())  
q.get()
print(q.full()) 
```

## qsize()

```python
q = queue.Queue()
q.put(10)
q.put(20)
print(q.qsize()) 
```

## PriorityQueue.put()

```python
q = queue.PriorityQueue()
q.put((3, 'task3'))
q.put((1, 'task1'))
print(q.queue) 
```

## PriorityQueue.get()

```python
q = queue.PriorityQueue()
q.put((3, 'task3'))
q.put((1, 'task1'))
item = q.get()
print(item) 
```

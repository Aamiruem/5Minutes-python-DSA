## 1. heapq.heappush()

```python
import heapq

heap = [1, 3, 5, 7, 9]
heapq.heapify(heap)

heapq.heappush(heap, 4) 
print(heap)
```

## 2. heapq.heappop()

```python
import heapq

heap = [1, 3, 5, 7, 9]
heapq.heapify(heap) 

smallest = heapq.heappop(heap)
print(smallest)  
print(heap)  
```

## 3. heapq.heappushpop()

```python
import heapq

heap = [1, 3, 5, 7, 9]
heapq.heapify(heap)

result = heapq.heappushpop(heap, 6)  
print(result)  
print(heap)  
```

## 4. heapq.heapreplace()

```python
import heapq

heap = [1, 3, 5, 7, 9]
heapq.heapify(heap)  

result = heapq.heapreplace(heap, 4) 
print(result)  
print(heap)
```

## 5. heapq.nlargest()

```python
import heapq

data = [1, 3, 5, 7, 9]
largest = heapq.nlargest(3, data) 
print(largest)  
```

## random.random()

```python
import random

result = random.random()
print(result)
```

## random.randint(a, b)

```python
result = random.randint(1, 10)
print(result)
```

## random.choice(seq)

```python
data = [10, 20, 30, 40, 50]
result = random.choice(data)
print(result)
```

## random.shuffle(seq)

```python
data = [1, 2, 3, 4, 5]
random.shuffle(data)
print(data)
```

## random.sample(seq, k)

```python
data = [1, 2, 3, 4, 5]
sample = random.sample(data, 3)
print(sample)
```

## random.seed(n)

```python
random.seed(10)
print(random.random())
```

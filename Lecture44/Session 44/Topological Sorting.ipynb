### Kahn’s Algorithm

```python
from collections import deque, defaultdict

def topo_sort_kahn(graph):
    indeg = defaultdict(int)
    for u in graph:
        indeg[u] = indeg[u]  
        for v in graph[u]:
            indeg[v] += 1

    q = deque([u for u in indeg if indeg[u] == 0])
    order = []

    while q:
        u = q.popleft()
        order.append(u)
        for v in graph[u]:
            indeg[v] -= 1
            if indeg[v] == 0:
                q.append(v)

    return order if len(order) == len(indeg) else None

graph_dag = {
    5: [2, 0],
    4: [0, 1],
    2: [3],
    3: [1],
    0: [],
    1: []
}

print("Kahn topology:", topo_sort_kahn(graph_dag))
```

### DFS-based Topological Sort

```python
def topo_sort_dfs(graph):
    visited = set()
    visiting = set() 
    order = []        

    def dfs(u):
        if u in visiting:   
            return False
        if u in visited:
            return True

        visiting.add(u)
        for v in graph[u]:
            if not dfs(v):
                return False
        visiting.remove(u)
        visited.add(u)
        order.append(u)
        return True

    for u in graph:
        if u not in visited:
            if not dfs(u):
                return None  

    return order[::-1]  

print("DFS topo:", topo_sort_dfs(graph_dag))
```

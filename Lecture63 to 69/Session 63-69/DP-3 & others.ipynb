```python
def count_subsets(arr, target):
    dp = [0] * (target + 1)
    dp[0] = 1  

    for num in arr:
        for s in range(target, num - 1, -1):
            dp[s] += dp[s - num]
    return dp[target]


arr = [1, 2, 3, 3]
target = 6
print(count_subsets(arr, target)) 
```

```python
def count_target_sum(nums, S):
    total = sum(nums)
    if (total + S) % 2 != 0 or abs(S) > total:
        return 0
    target = (total + S) // 2
 
    dp = [0] * (target + 1)
    dp[0] = 1
    for num in nums:
        for s in range(target, num - 1, -1):
            dp[s] += dp[s - num]
    return dp[target]


nums = [1,1,1,1,1]
S = 3
print(count_target_sum(nums, S)) 
```

```python
def count_k_sum(arr, k, target):
    dp = [[0] * (target + 1) for _ in range(k + 1)]
    dp[0][0] = 1

    for num in arr:
        for j in range(k, 0, -1):
            for s in range(target, num - 1, -1):
                dp[j][s] += dp[j - 1][s - num]
    return dp[k][target]


arr = [1,2,3,4,5]
k = 2
target = 6
print(count_k_sum(arr, k, target))  
```

```python
def subtree_sizes(root, graph):
    n = len(graph)
    size = {u:0 for u in graph}
    def dfs(u, p):
        size[u] = 1
        for v in graph[u]:
            if v == p: continue
            dfs(v, u)
            size[u] += size[v]
    dfs(root, -1)
    return size



graph = {1:[2,3], 2:[1,4], 3:[1], 4:[2]}
print(subtree_sizes(1, graph)) 
```

```python
graph = {
    1: [2, 3],
    2: [1, 4, 5],
    3: [1],
    4: [2],
    5: [2]
}

diameter = 0

def dfs(node, parent):
    global diameter
    max1 = 0
    max2 = 0

    for child in graph[node]:
        if child == parent:
            continue
        height = dfs(child, node)

        if height > max1:
            max2 = max1
            max1 = height
        elif height > max2:
            max2 = height

    diameter = max(diameter, max1 + max2)

    return max1 + 1


dfs(1, -1)
print("Diameter of tree (in edges):", diameter)
```

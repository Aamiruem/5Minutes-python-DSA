## Longest Common Subsequence (LCS)

```python
def lcs(a,b):
    n,m = len(a), len(b)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for j in range(1,m+1):
            if a[i-1]==b[j-1]:
                dp[i][j] = dp[i-1][j-1]+1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    i,j = n,m
    res = []
    while i>0 and j>0:
        if a[i-1]==b[j-1]:
            res.append(a[i-1]); i-=1; j-=1
        elif dp[i-1][j] >= dp[i][j-1]:
            i-=1
        else:
            j-=1
    return dp[n][m], "".join(reversed(res))

print(lcs("AGGTAB","GXTXAYB"))
```

## Edit Distance / Levenshtein Distance (insert, delete, replace)

```python
def edit_distance(a,b):
    n,m = len(a), len(b)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(n+1): dp[i][0] = i
    for j in range(m+1): dp[0][j] = j
    for i in range(1,n+1):
        for j in range(1,m+1):
            if a[i-1]==b[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1])
    return dp[n][m]

print(edit_distance("kitten","sitting"))
```

## Palindrome Substrings

```python
def pal_substrings(s):
    n = len(s)
    dp = [[False]*n for _ in range(n)]
    count = 0
    best_len = 0; best_i = 0
    for length in range(1, n+1):
        for i in range(0, n-length+1):
            j = i+length-1
            if s[i]==s[j] and (length<=2 or dp[i+1][j-1]):
                dp[i][j] = True
                count += 1
                if length > best_len:
                    best_len = length; best_i = i
    longest = s[best_i:best_i+best_len]
    return count, longest

print(pal_substrings("ababa")) 
```

## 0/1 Knapsack (Recursive)

```python
def knap_rec(wt,val,W,n):
    if n==0 or W==0: return 0
    if wt[n-1] > W:
        return knap_rec(wt,val,W,n-1)
    take = val[n-1] + knap_rec(wt,val,W-wt[n-1],n-1)
    skip = knap_rec(wt,val,W,n-1)
    return max(take,skip)

wt=[1,2,3]
val=[6,10,12]
W=5
print(knap_rec(wt,val,W,len(wt))) 
```

## 0/1 Knapsack (Memoized – Top Down)

```python
def knap_memo(wt,val,W,n,memo={}):
    if (n,W) in memo: return memo[(n,W)]
    if n==0 or W==0: return 0
    if wt[n-1] > W:
        memo[(n,W)] = knap_memo(wt,val,W,n-1,memo)
    else:
        take = val[n-1] + knap_memo(wt,val,W-wt[n-1],n-1,memo)
        skip = knap_memo(wt,val,W,n-1,memo)
        memo[(n,W)] = max(take,skip)
    return memo[(n,W)]

print(knap_memo([1,2,3],[6,10,12],5,3))  
```

## 0/1 Knapsack (Tabulation – Bottom Up)

```python
def knap_tab(wt,val,W,n):
    dp=[[0]*(W+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for w in range(1,W+1):
            if wt[i-1] <= w:
                dp[i][w] = max(val[i-1]+dp[i-1][w-wt[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]
    return dp[n][W]

print(knap_tab([1,2,3],[6,10,12],5,3)) 
```

## Recursive(mcm)

```python
def mcm_rec(d, i, j):
    if i == j:
        return 0
    best = 10**18
    for k in range(i, j):
        cost = mcm_rec(d, i, k) + mcm_rec(d, k+1, j) + d[i-1]*d[k]*d[j]
        if cost < best:
            best = cost
    return best

dims = [40,20,30,10,30]
n = len(dims)-1
print(mcm_rec(dims, 1, n)) 
```

## Memoized (top-down)

```python
def mcm_memo(d):
    n = len(d)-1
    memo = {}
    def helper(i,j):
        if i==j: return 0
        if (i,j) in memo: return memo[(i,j)]
        best = 10**18
        for k in range(i,j):
            cost = helper(i,k) + helper(k+1,j) + d[i-1]*d[k]*d[j]
            if cost < best:
                best = cost
        memo[(i,j)] = best
        return best
    return helper(1,n)

dims = [40,20,30,10,30]
print(mcm_memo(dims))  
```

## Longest Palindromic Subsequence (LPS)

```python
def lcs_table(a,b):
    n,m = len(a), len(b)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for j in range(1,m+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp

def lcs_from_table(a,b,dp):
    i,j = len(a), len(b)
    res = []
    while i>0 and j>0:
        if a[i-1] == b[j-1]:
            res.append(a[i-1]); i-=1; j-=1
        elif dp[i-1][j] >= dp[i][j-1]:
            i-=1
        else:
            j-=1
    return "".join(reversed(res))

def lps(s):
    rev = s[::-1]
    dp = lcs_table(s, rev)
    subseq = lcs_from_table(s, rev, dp)
    return dp[len(s)][len(s)], subseq

print(lps("bbbab")) 
```

## Minimum insertions / deletions to convert str1 → str2

```python
def min_ins_del(s1, s2):
    dp = lcs_table(s1, s2)
    l = dp[len(s1)][len(s2)]
    deletions = len(s1) - l
    insertions = len(s2) - l
    return deletions, insertions

print(min_ins_del("heap", "pea"))  
```

## Shortest Common Supersequence (SCS)

```python
def scs(a,b):
    dp = lcs_table(a,b)
    i, j = len(a), len(b)
    res = []
    while i>0 and j>0:
        if a[i-1] == b[j-1]:
            res.append(a[i-1]); i-=1; j-=1
        elif dp[i-1][j] >= dp[i][j-1]:
            res.append(a[i-1]); i-=1
        else:
            res.append(b[j-1]); j-=1
    while i>0:
        res.append(a[i-1]); i-=1
    while j>0:
        res.append(b[j-1]); j-=1
    return "".join(reversed(res))

print(scs("abac", "cab"))  
```

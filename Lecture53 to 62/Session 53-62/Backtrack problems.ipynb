```python
def solve_n_queens(n):
    ans = []
    board = [["."]*n for _ in range(n)]
    
    print(board)
    
    def safe(row, col):
        for i in range(row):
            if board[i][col] == "Q":
                return False
        i, j = row-1, col-1
        while i >= 0 and j >= 0:
            if board[i][j] == "Q": return False
            i, j = i-1, j-1
        i, j = row-1, col+1
        while i >= 0 and j < n:
            if board[i][j] == "Q": return False
            i, j = i-1, j+1
        return True
    
    def backtrack(row):
        if row == n:
            ans.append(["".join(r) for r in board])
            print(ans)
            return
        for col in range(n):
            if safe(row, col):
                board[row][col] = "Q"
                backtrack(row+1)
                board[row][col] = "."
    
    backtrack(0)
    return ans

print(len(solve_n_queens(4)))
```

```python
def rat_maze_paths(grid):
    R = len(grid)
    C = len(grid[0])
    visited = [[False]*C for _ in range(R)]
    paths = []
    cur = []

    def inside(r,c):
        return 0 <= r < R and 0 <= c < C

    def dfs(r,c):
        if not inside(r,c) or grid[r][c] == 0 or visited[r][c]:
            return
        visited[r][c] = True
        cur.append((r,c))
        if r == R-1 and c == C-1:
            paths.append(cur[:])
        else:
            dfs(r+1,c)
            dfs(r,c+1)
            dfs(r-1,c)
            dfs(r,c-1)
        cur.pop()
        visited[r][c] = False

    dfs(0,0)
    return paths

grid = [[1,0,1,1],
        [1,1,1,0],
        [0,1,1,1],
        [0,0,1,1]]

print(len(rat_maze_paths(grid)))
```

```python
def solve_sudoku_4x4(board):
    n = 4
    bs = 2

    def find_empty():
        for r in range(n):
            for c in range(n):
                if board[r][c] == 0:
                    return r, c
        return None

    def valid(r, c, v):
        for i in range(n):
            if board[r][i] == v or board[i][c] == v:
                return False
        br, bc = (r//bs)*bs, (c//bs)*bs
        for i in range(br, br+bs):
            for j in range(bc, bc+bs):
                if board[i][j] == v:
                    return False
        return True

    loc = find_empty()
    if not loc:
        return True
    r, c = loc
    for v in range(1, n+1):
        if valid(r, c, v):
            board[r][c] = v
            print(v)
            if solve_sudoku_4x4(board):
                return True
            board[r][c] = 0
    return False

puzzle = [
    [1,0,0,3],
    [0,0,1,0],
    [0,3,0,0],
    [2,0,0,4]
]

solve_sudoku_4x4(puzzle)
for row in puzzle:
    print(row)
```

```python
def exist(board, word):
    R = len(board)
    C = len(board[0])
    visited = [[False]*C for _ in range(R)]

    def inside(r,c):
        return 0 <= r < R and 0 <= c < C

    def dfs(r,c,i):
        if i == len(word):
            return True
        if not inside(r,c) or visited[r][c] or board[r][c] != word[i]:
            return False
        visited[r][c] = True
        if (dfs(r+1,c,i+1) or dfs(r-1,c,i+1) or
            dfs(r,c+1,i+1) or dfs(r,c-1,i+1)):
            return True
        visited[r][c] = False
        return False

    for r in range(R):
        for c in range(C):
            if dfs(r,c,0):
                return True
    return False

b = [list("ABCE"), list("SFCS"), list("ADEE")]
print(exist(b, "ABCCED")) 
print(exist(b, "SEE"))    
print(exist(b, "ABCB"))    
```

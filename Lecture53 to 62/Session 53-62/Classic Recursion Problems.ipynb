```python
def fact(n):
    if n == 0: return 1
    return n * fact(n-1)

print(fact(5)) 
```

```python
def fib(n):
    if n <= 1: return n
    return fib(n-1) + fib(n-2)

print([fib(i) for i in range(7)])  
```

```python
def sumn(n):
    if n == 0: return 0
    return n + sumn(n-1)

print(sumn(5)) 
```

```python
def rev(s):
    if s == "": return ""
    return rev(s[1:]) + s[0]

print(rev("hello"))
```

```python
def count_digits(n):
    if n == 0: return 0
    return 1 + count_digits(n//10)

print(count_digits(12345)) 
```

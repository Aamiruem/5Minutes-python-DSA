## Recursion (no DP)

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(5))
```

## Memoization (Top-Down DP)

```python
def fib_memo(n, memo={}):
    if n in memo: 
        return memo[n]
    if n <= 1:
        memo[n] = n
    else:
        memo[n] = fib_memo(n-1,memo) + fib_memo(n-2,memo)
    return memo[n]

print(fib_memo(5)) 
```

## Tabulation (Bottom-Up DP)

```python
def fib_tab(n):
    dp = [0,1]
    for i in range(2,n+1):
        dp.append(dp[i-1]+dp[i-2])
    return dp[n]

print(fib_tab(5))
```

### Overlapping Subproblems
fib(5) needs fib(4) and fib(3)...
fib(4) again needs fib(3) and fib(2)...
fib(3) is solved multiple times (overlap)

### Optimal Substructure

```python
def min_path_sum(grid):
    R,C = len(grid), len(grid[0])
    dp = [[0]*C for _ in range(R)]
    dp[0][0] = grid[0][0]
    for r in range(R):
        for c in range(C):
            if r==0 and c==0: continue
            up = dp[r-1][c] if r>0 else float('inf')
            left = dp[r][c-1] if c>0 else float('inf')
            dp[r][c] = grid[r][c] + min(up,left)
    return dp[R-1][C-1]

g = [[1,3,1],[1,5,1],[4,2,1]]
print(min_path_sum(g)) 
```

### Manual Memoization with Dictionary

```python
def fib(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        memo[n] = n
    else:
        memo[n] = fib(n-1, memo) + fib(n-2, memo)
    return memo[n]

print(fib(6))  
```

## Manual Memoization with Array

```python
def fib_arr(n):
    memo = [-1]*(n+1)
    def helper(k):
        if memo[k] != -1:
            return memo[k]
        if k <= 1:
            memo[k] = k
        else:
            memo[k] = helper(k-1) + helper(k-2)
        return memo[k]
    return helper(n)

print(fib_arr(6)) 
```

### Avoiding Repeated Work

```python
def fib_plain(n):
    if n <= 1:
        return n
    return fib_plain(n-1) + fib_plain(n-2)

print(fib_plain(6))
```

## Iterative Approach (Bottom-Up DP)

```python
def fib_tab(n):
    dp = [0]*(n+1)
    dp[0], dp[1] = 0, 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

print(fib_tab(6)) 
```

## Creating and Filling a DP Table

```python
def min_path_sum(grid):
    R, C = len(grid), len(grid[0])
    dp = [[0]*C for _ in range(R)]
    dp[0][0] = grid[0][0]
    for r in range(R):
        for c in range(C):
            if r==0 and c==0: continue
            up = dp[r-1][c] if r>0 else float('inf')
            left = dp[r][c-1] if c>0 else float('inf')
            dp[r][c] = grid[r][c] + min(up, left)
    return dp[R-1][C-1]

g = [[1,3,1],[1,5,1],[4,2,1]]
print(min_path_sum(g)) 
```

## Converting Recursion to Tabulation

```python
def fact_rec(n):
    if n<=1: return 1
    return n * fact_rec(n-1)

def fact_tab(n):
    dp = [1]*(n+1)
    for i in range(2, n+1):
        dp[i] = i * dp[i-1]
    return dp[n]

print(fact_tab(5)) 
```

## Space Optimization (2D → 1D)

```python
def count_paths(R, C):
    dp = [1]*C
    for _ in range(1,R):
        for c in range(1,C):
            dp[c] += dp[c-1]
    return dp[-1]

print(count_paths(3,3)) 
```

## Reusing Previous States (Rolling Variables)

```python
def fib_opt(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a+b
    return a

print(fib_opt(6))
```

## Fibonacci Number

```python
def fib(n):
    if n <= 1: return n
    dp = [0]*(n+1)
    dp[0], dp[1] = 0, 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

print(fib(6)) 
```

## Climbing Stairs

```python
def climb_stairs(n):
    dp = [0]*(n+1)
    dp[0], dp[1] = 1, 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

print(climb_stairs(5))  
```

## Minimum Cost Climbing Stairs

```python
def min_cost(cost):
    n = len(cost)
    dp = [0]*(n+1)
    dp[0], dp[1] = 0, 0
    for i in range(2, n+1):
        dp[i] = min(dp[i-1]+cost[i-1], dp[i-2]+cost[i-2])
    return dp[n]

print(min_cost([10,15,20])) 
```

## House Robber

```python
def rob(nums):
    if not nums: return 0
    if len(nums)==1: return nums[0]
    dp = [0]*len(nums)
    dp[0], dp[1] = nums[0], max(nums[0],nums[1])
    for i in range(2,len(nums)):
        dp[i] = max(dp[i-1], dp[i-2]+nums[i])
    return dp[-1]

print(rob([2,7,9,3,1])) 
```

## Jump Game

```python
def can_jump(nums):
    far = 0
    for i,x in enumerate(nums):
        if i > far: return False
        far = max(far, i+x)
    return True

print(can_jump([2,3,1,1,4])) 
print(can_jump([3,2,1,0,4]))  
```

## Unique paths

```python
def unique_paths(r, c):
    dp = [[0]*c for _ in range(r)]
    for i in range(r):
        dp[i][0] = 1
    for j in range(c):
        dp[0][j] = 1
    for i in range(1, r):
        for j in range(1, c):
            dp[i][j] = dp[i-1][j] + dp[i][j-1]
    return dp[-1][-1]

print(unique_paths(3, 3)) 
```

## Unique paths with obstacles

```python
def unique_paths_obstacle(grid):
    R, C = len(grid), len(grid[0])
    dp = [[0]*C for _ in range(R)]
    if grid[0][0] == 0:
        dp[0][0] = 1
    for i in range(1, R):
        if grid[i][0] == 0: dp[i][0] = dp[i-1][0]
    for j in range(1, C):
        if grid[0][j] == 0: dp[0][j] = dp[0][j-1]
    for i in range(1, R):
        for j in range(1, C):
            if grid[i][j] == 0:
                dp[i][j] = dp[i-1][j] + dp[i][j-1]
    return dp[-1][-1]

g = [[0,0,0],[0,1,0],[0,0,0]]
print(unique_paths_obstacle(g)) 
```

## Min path sum in a grid (from (0,0) to (n-1,m-1))

```python
def min_path_sum(grid):
    R, C = len(grid), len(grid[0])
    dp = [[0]*C for _ in range(R)]
    dp[0][0] = grid[0][0]
    for j in range(1, C):
        dp[0][j] = dp[0][j-1] + grid[0][j]
    for i in range(1, R):
        dp[i][0] = dp[i-1][0] + grid[i][0]
    for i in range(1, R):
        for j in range(1, C):
            dp[i][j] = grid[i][j] + min(dp[i-1][j], dp[i][j-1])
    return dp[-1][-1]

g = [[1,3,1],
     [1,5,1],
     [4,2,1]]
print(min_path_sum(g)) 
```

## Visualizing 2D DP with table + directions (for min path sum)

```python
def visualize_min_path(grid):
    R, C = len(grid), len(grid[0])
    dp = [[0]*C for _ in range(R)]
    dir = [['']*C for _ in range(R)]
    dp[0][0] = grid[0][0]
    for j in range(1, C):
        dp[0][j] = dp[0][j-1] + grid[0][j]; dir[0][j] = 'R'
    for i in range(1, R):
        dp[i][0] = dp[i-1][0] + grid[i][0]; dir[i][0] = 'D'
    for i in range(1, R):
        for j in range(1, C):
            if dp[i-1][j] < dp[i][j-1]:
                dp[i][j] = grid[i][j] + dp[i-1][j]; dir[i][j] = 'D'
            else:
                dp[i][j] = grid[i][j] + dp[i][j-1]; dir[i][j] = 'R'
    print("DP table:")
    for row in dp:
        print(row)
    i, j = R-1, C-1
    path = []
    while not (i==0 and j==0):
        path.append((i,j))
        if dir[i][j] == 'D':
            i -= 1
        else:
            j -= 1
    path.append((0,0))
    path.reverse()
    print("Path coords:", path)
    print("Directions:")
    for row in dir:
        print(row)

g = [[1,3,1],[1,5,1],[4,2,1]]
visualize_min_path(g)
```

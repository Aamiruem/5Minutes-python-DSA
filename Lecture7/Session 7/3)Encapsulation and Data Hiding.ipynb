## 1. Access Modifiers in Python
| Access Modifier | Syntax       | Access Level             | Accessible From   |
| --------------- | ------------ | ------------------------ | ----------------- |
| **Public**      | `variable`   | Fully accessible         | Anywhere          |
| **Protected**   | `_variable`  | Convention: internal use | Class + Subclass  |
| **Private**     | `__variable` | Name mangled             | Only inside class |


```python
class Sample:
    def __init__(self):
        self.public = "I'm public"
        self._protected = "I'm protected"
        self.__private = "I'm private"

s = Sample()
print(s.public)
print(s._protected)
# print(s.__private)
```

```python
class Sample:
    def __init__(self):
        self.variable = "I'm public"
        self._variable = "I'm protected"
        self.__variable = "I'm private"

s = Sample()
print(s.variable)
print(s._variable)
# print(s.__variable)
```

```python
print(s.__variable)
```

```python
print(s._Sample__variable)
```

## 2. Getters and Setters
* Allow controlled access to private/protected attributes.

```python
class Account:
    def __init__(self):
        self.__balance = 0

    def get_balance(self):
        return self.__balance

    def set_balance(self, amount):
        if amount >= 0:
            self.__balance = amount
        else:
            print("Invalid balance!")

acc = Account()
acc.set_balance(1000)
print(acc.get_balance())
```

## 3. Property Decorators
* Pythonic way to use getters/setters.
* Use `@property` to turn a method into a getter.
* Use `@<property>.setter` for controlled setting.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value > 0:
            self._radius = value
        else:
            print("Radius must be positive!")

c = Circle(5)
print(c.radius)
c.radius = 10
print(c.radius)
```

```python
c.radius = -10
```

## 4. Hiding Internal Object Details
* You can use `__` to indicate internals not meant for external use.
* Useful for preventing accidental modification or enforcing constraints.

```python
class Engine:
    def __init__(self):
        self.__temperature = 75

    def run(self):
        self.__temperature += 10
        print("Engine running. Temp:", self.__temperature)

e = Engine()
e.run()
print(e._Engine__temperature)
```

```python
e.__temperature = 500
print(e.__temperature)
```

```python
e.run()
```

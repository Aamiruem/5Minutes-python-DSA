## 1. Instance Methods
* Most common type of method.
* Takes `self` as the first argument.
* Can access and modify **object-specific data**.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, my name is {self.name}."

p = Person("Shridhar")
print(p.greet())
```

## 2. Class Methods
* Use the `@classmethod` decorator.
* First argument is `cls`, not `self`.
* Can access/modify **class-level attributes**.

```python
class Book:
    library_name = "City Library"

    @classmethod
    def get_library(cls):
        return f"Books are stored at {cls.library_name}."

print(Book.get_library())
```

```python
class Book1:
    name = "Shridhar"
    
    def __init__(self, title):
        self.title = title

    @classmethod
    def get_name(cls):
        print(cls.title)

b1 = Book1("ABC")
b1.get_name()
```

## 3. Static Methods
* Use the `@staticmethod` decorator.
* Takes **no special first argument** (`self` or `cls`).
* Behaves like a regular function but lives inside the class.

```python
class Math:
    @staticmethod
    def add(x, y):
        return x + y


print(Math.add(10, 20))
```

```python
m = Math() 
print(m.add(100, 20))
```

## 4. When and Why to Use Each

| Method Type     | First Param | Access `self`? | Access `cls`? | Use When                      |
| --------------- | ----------- | -------------- | ------------- | -------------------------------- |
| Instance Method | `self`      | Yes              | No             | Needs object-specific data       |
| Class Method    | `cls`       | No              | Yes             | Needs class-level data           |
| Static Method   | None        | No              | No             | Pure function grouped in a class |


```python
class Math1:
    def add(self, x, y):
        return x + y
```

```python
m1 = Math1() 
print(m1.add(100, 20))
```

```python
print(Math1.add(10, 20))
```

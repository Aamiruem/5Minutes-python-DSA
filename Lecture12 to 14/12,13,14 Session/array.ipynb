## 2. array.append()

```python
import array

arr = array.array('i', [1, 2, 3])
arr.append(4)
print(arr)
```

## 3. array.extend()

```python
import array

arr = array.array('i', [1, 2, 3])
arr.extend([4, 5, 6])
print(arr) 
```

## 4. array.insert()

```python
import array

arr = array.array('i', [1, 2, 3, 5])
arr.insert(3, 4) 
print(arr)
```

## 5. array.remove()

```python
import array

arr = array.array('i', [1, 2, 3, 4, 3, 5])
arr.remove(3)
print(arr)  
```

## 6. array.pop()

```python
import array

arr = array.array('i', [1, 2, 3, 4, 5])
popped = arr.pop(2) 
print(popped)  
print(arr) 
```

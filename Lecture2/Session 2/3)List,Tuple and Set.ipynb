# This error means you tried to use .append() on a tuple, but tuples in Python are immutable (unchangeable).

Example:

tuple2 = (1, 2, 3, 4)
tuple2.append("five")   # ❌ AttributeError

Why?

Lists ([]) are mutable → they support methods like .append(), .remove(), .pop().

Tuples (()) are immutable → once created, you cannot add, remove, or change elements.

✅ Correct ways to add "five":

Use a list instead of a tuple

list1 = [1, 2, 3, 4]
list1.append("five")
print(list1)   # [1, 2, 3, 4, 'five']


Create a new tuple (since tuples can’t be changed)

tuple2 = (1, 2, 3, 4)
tuple2 = tuple2 + ("five",)
print(tuple2)   # (1, 2, 3, 4, 'five')


👉 In short:
You can’t append to a tuple — you must either use a list or create a new tuple.

## This error means Python has no idea what tuple2 is, because you never defined it before trying to use it.

For example:

print(tuple2)   # ❌ NameError: name 'tuple2' is not defined

✅ How to fix

You need to create (define) tuple2 before using it:

tuple2 = (1, 2, 3, 4)
print(tuple2)

Why it happens

A NameError means you used a variable name that doesn’t exist yet.

It could be a typo, or you forgot to assign a value before using it.

👉 In short:
You must define tuple2 first before you can access it.

# This error means you’re trying to access a set element using an index ([0]), but Python sets do not support indexing.

For example:

set4 = {10, 20, 30}
print(set4[0])   # ❌ This will throw "TypeError: 'set' object is not subscriptable"


Why?

Lists ([]) and tuples (()) are ordered collections, so you can access elements by index.

Sets ({}) are unordered collections, so they don’t guarantee order, and you can’t use indexing.

✅ Correct ways to work with sets:

Loop through the set

for item in set4:
    print(item)


Convert to a list if you need indexing

set4 = {10, 20, 30}
print(list(set4)[0])  # Works, but order is arbitrary


Use set operations (preferred)

print(20 in set4)  # Check membership


👉 In short:
'set' object is not subscriptable means "you can’t use [index] on a set".

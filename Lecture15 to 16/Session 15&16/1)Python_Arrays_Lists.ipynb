# Arrays / Lists in Python

## 1. Python Lists as Dynamic Arrays

```python
arr = [1, 2, 3]
arr.append(4)
print(arr)
```

## 2. Creating and Initializing Lists

```python
a = []
b = [1, 2, 3]
c = list(range(5))  
d = [0] * 5 
print(a, b, c, d)
```

## 3. Indexing and Slicing

```python
lst = [10, 20, 30, 40, 50]
print(lst[0])   
print(lst[-1])  
print(lst[1:4])  
print(lst[:3])  
print(lst[::2])  
```

## 4. List Methods

```python

lst = [1, 2]
lst.append(3)
print("append:", lst)

lst.extend([4, 5])
print("extend:", lst)

lst.insert(2, 99)
print("insert:", lst)

lst.remove(99)
print("remove:", lst)

popped = lst.pop()
print("pop:", popped, "list:", lst)

print("index of 2:", lst.index(2))
print("count of 2:", lst.count(2))

lst.sort()
print("sort:", lst)

lst.reverse()
print("reverse:", lst)

copied = lst.copy()
copied.append(100)
print("copy original:", lst)
print("copy modified:", copied)
```

## 5. List Comprehensions

```python
squares = [x**2 for x in range(5)]
print("Squares:", squares)

evens = [x for x in range(10) if x % 2 == 0]
print("Evens:", evens)
```

## 6. Nested Lists and 2D Arrays

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("Element at [1][2]:", matrix[1][2])

print("Row-wise traversal:")
for row in matrix:
    print(row)

print("Element-wise traversal:")
for row in matrix:
    for col in row:
        print(col, end=' ')
print()
```

## 7. Time and Space Complexity of List Operations

| Operation          | Time Complexity | Space Complexity |
|--------------------|------------------|------------------|
| Indexing           | O(1)             | O(1)             |
| Append             | O(1)             | O(1)             |
| Insert at position | O(n)             | O(1)             |
| Delete at position | O(n)             | O(1)             |
| Search (linear)    | O(n)             | O(1)             |
| Sort               | O(n log n)       | O(n)             |
| Copy               | O(n)             | O(n)             |

```python
grid = [
    [1,0,1],
    [1,1,0],
    [0,1,1]
]
R, C = len(grid), len(grid[0])
print("Grid:", grid)
print("Rows:", R, "Cols:", C)
```

### Convert 2D grid to graph

```python
def grid_to_graph(g):
    R, C = len(g), len(g[0])
    adj = {r*C+c:[] for r in range(R) for c in range(C) if g[r][c]}
    for r in range(R):
        for c in range(C):
            if not g[r][c]: 
                continue
            i = r*C+c
            for dr,dc in [(1,0),(-1,0),(0,1),(0,-1)]:
                nr,nc = r+dr, c+dc
                if 0<=nr<R and 0<=nc<C and g[nr][nc]:
                    adj[i].append(nr*C+nc)
    return adj

graph = grid_to_graph(grid)
print("Graph:", graph)
```

### BFS on grid

```python
from collections import deque

def bfs_grid(g, sr, sc):
    R, C = len(g), len(g[0])
    if g[sr][sc] == 0: 
        return []
    
    q = deque([(sr, sc)])
    vis = {(sr, sc)}   
    order = []
    
    while q:
        r, c = q.popleft()
        order.append((r, c))
        for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]:
            nr, nc = r+dr, c+dc
            if (0 <= nr < R and 0 <= nc < C 
                and g[nr][nc] == 1 
                and (nr, nc) not in vis):
                vis.add((nr, nc))
                q.append((nr, nc))
    return order


grid = [
    [1,0,1],
    [1,1,0],
    [0,1,1]
]

print("BFS:", bfs_grid(grid, 0, 0))
```

### DFS on grid

```python
def dfs_grid(g, sr, sc):
    R, C = len(g), len(g[0])
    if g[sr][sc] == 0:
        return []
    
    vis = set()
    order = []
    
    def dfs(r, c):
        vis.add((r, c))
        order.append((r, c))
        for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]:
            nr, nc = r+dr, c+dc
            if (0 <= nr < R and 0 <= nc < C 
                and g[nr][nc] == 1 
                and (nr, nc) not in vis):
                dfs(nr, nc)
    
    dfs(sr, sc)
    return order


grid = [
    [1,0,1],
    [1,1,0],
    [0,1,1]
]

print("DFS:", dfs_grid(grid, 0, 0))
```

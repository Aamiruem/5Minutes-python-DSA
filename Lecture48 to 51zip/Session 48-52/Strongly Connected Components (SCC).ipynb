```python
graph = {
    0:[1],
    1:[2],
    2:[0,3],
    3:[4],
    4:[5,7],
    5:[6],
    6:[4],
    7:[8],
    8:[7]
}
print("Graph:", graph)
```

### Kosaraju’s Algorithm

```python
from collections import defaultdict

def kosaraju(graph):
    n = len(graph)
    vis = set()
    order = []

    def dfs(u):
        vis.add(u)
        print("DFS-1 Visit",u)
        for v in graph[u]:
            if v not in vis:
                dfs(v)
        order.append(u)
        print("Finish at -->",u)

    for u in graph:
        if u not in vis:
            dfs(u)

    rev = defaultdict(list)
    for u in graph:
        for v in graph[u]:
            rev[v].append(u)
            print(f"Reversed edge {v} <--> {u}")

    vis.clear()
    sccs = []
    def dfs_rev(u, comp):
        vis.add(u)
        comp.append(u)
        for v in rev[u]:
            if v not in vis:
                dfs_rev(v, comp)

    print("DFS-1 Order = ",order)
    for u in reversed(order):
        if u not in vis:
            comp = []
            dfs_rev(u, comp)
            sccs.append(comp)
    return sccs

print("Kosaraju SCC:", kosaraju(graph))
```

### Tarjan’s Algorithm

```python
def tarjan(graph):
    n = len(graph)
    index = {}
    low = {}
    stack = []
    onstack = set()
    res = []
    idx = [0]

    def dfs(u):
        index[u] = low[u] = idx[0]
        idx[0] += 1
        stack.append(u)
        onstack.add(u)
        for v in graph[u]:
            if v not in index:
                dfs(v)
                low[u] = min(low[u], low[v])
            elif v in onstack:
                low[u] = min(low[u], index[v])
        if low[u] == index[u]:
            comp = []
            while True:
                w = stack.pop()
                onstack.remove(w)
                comp.append(w)
                if w == u:
                    break
            res.append(comp)

    for u in graph:
        if u not in index:
            dfs(u)
    return res

print("Tarjan SCC:", tarjan(graph))
```

```python
def make_set(n):
    parent = [i for i in range(n)]
    rank = [0] * n
    return parent, rank
```

```python
def find(x,p):
    if p[x]!=x: p[x]=find(p[x],p)
    return p[x]
```

```python
def union(a,b,p,r):
    a,b=find(a,p),find(b,p)
    if a==b: return False
    if r[a]<r[b]: p[a]=b
    elif r[a]>r[b]: p[b]=a
    else: p[b]=a; r[a]+=1
    return True
```

```python
parent, rank = make_set(5)

print("Initial parent:", parent)
print("Initial rank:", rank)

union(0, 1, parent, rank)
print("\nAfter union(0,1):")
print("Parent:", parent, "Rank:", rank)

union(1, 2, parent, rank)
print("\nAfter union(1,2):")
print("Parent:", parent, "Rank:", rank)

union(3, 4, parent, rank)
print("\nAfter union(3,4):")
print("Parent:", parent, "Rank:", rank)

union(2, 4, parent, rank)
print("\nAfter union(2,4):")
print("Parent:", parent, "Rank:", rank)

print("\nFind(2):", find(2, parent))
print("Find(4):", find(4, parent))
print("Final Parent:", parent)
```

```python
def has_cycle_undirected(n, edges):
    parent, rank = make_set(n)
    for u, v in edges:
        if not union(u, v, parent, rank):  
            return True
    return False
```

```python
def connected_components(n, edges):
    parent, rank = make_set(n)
    for u, v in edges:
        union(u, v, parent, rank)
    comps = {}
    for i in range(n):
        r = find(i, parent)
        comps.setdefault(r, []).append(i)
    return list(comps.values())
```

```python
def kruskal(n, edges_w):
    p, r = make_set(n)
    cost, mst = 0, []
    for u, v, w in sorted(edges_w, key=lambda e: e[2]):
        if u == v: 
            continue
        if union(u, v, p, r):
            mst.append((u, v, w))
            cost += w
    return cost, mst

edges_w = [
    (0, 1, 10),
    (0, 2, 6),
    (0, 3, 5),
    (1, 3, 15),
    (2, 3, 4)
]
n = 4
print("MST:", kruskal(n, edges_w))
```

```python
edges_unweighted = [(0,1), (1,2), (2,0), (3,4)]
print("Cycle exists:", has_cycle_undirected(5, edges_unweighted)) 

print("Connected components:", connected_components(5, [(0,1), (1,2), (3,4)])) 
```

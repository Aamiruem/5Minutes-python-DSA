```python
dict3 = {
    "sports_id": 1,
    "sports_name": "cricket",
    "score1": [20, 40, 70, 2],
    "score2": (1, 2, 3, 4, 5),
    "score3": {10, 20, 30, 40, 50},
}
```

```python
dict3
```

```python
dict4 = dict(product_id=1, product_name="xyz", product_price=999)
```

```python
dict4
```

```python
dict3.keys()
```

```python
dict3["win"] = True
```

```python
dict3.keys()
```

```python
dict3.values()
```

```python
dict3["win"] = False
```

```python
dict3.values() # This will update the value of the key "win" to False. The keys of the dictionary will remain the same, but the value associated with the key "win" will be updated to False.
```

```python
dict3.items() 
```

```python
dict3.update({"win": True}) # This will update the value of the key "win" to True. If the key "win" does not exist in the dictionary, it will be added with the value True.
```

```python
dict4 = dict3.copy() # This will create a new dictionary dict4 that is a copy of dict3. Any changes made to dict4 will not affect dict3, and vice versa.
dict4
```

```python
dict3
```

```python
dict3.pop("win") # This will remove the key "win" and its associated value from the dictionary and return the value that was removed. In this case, it will return True.
```

```python
dict3
```

```python
dict3.popitem() # This will remove and return the last inserted key-value pair from the dictionary. In this case, it will remove the key "score3" and its associated value.
```

```python
dict3
```

```python
del dict3["score2"] # This will delete the key "score2" and its associated value from the dictionary.
```

```python
dict3
```

```python
dict3.clear() # This will remove all the key-value pairs from the dictionary, leaving it empty.
```

```python
dict3
```

```python
del dict3  # This will delete the entire dictionary and we won't be able to access it anymore.
```

```python
dict3  # This will raise an error because dict3 has been deleted and is no longer accessible.  # noqa: F821
```

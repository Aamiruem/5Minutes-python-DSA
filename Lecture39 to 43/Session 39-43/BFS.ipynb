### BFS using Queue

```python
from collections import deque

def bfs(start, graph):
    visited = set()
    queue = deque([start])   
    
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            print(node, end=" ")
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

graph = {
    0: [1, 2],
    1: [0, 3],
    2: [0, 3],
    3: [1, 2]
}

bfs(0, graph)
```

### Level-order Traversal (in Unweighted Graphs)

```python
def bfs_levels(start, graph):
    visited = set([start])
    queue = deque([(start, 0)])  
    
    while queue:
        node, level = queue.popleft()
        print(f"Node {node} at level {level}")
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, level+1))

bfs_levels(0, graph)
```

### Applications of BFS
Shortest Path in Unweighted Graphs

```python
def shortest_path(start, end, graph):
    visited = set([start])
    queue = deque([(start, [start])])   
    
    while queue:
        node, path = queue.popleft()
        if node == end:
            return path
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None

print("Shortest Path from 0 to 3:", shortest_path(0, 3, graph))
```

Checking Bipartite Graphs

```python
def is_bipartite(graph):
    color = {}
    for start in graph:
        if start not in color:
            queue = deque([start])
            color[start] = 0 
            
            while queue:
                node = queue.popleft()
                for neighbor in graph[node]:
                    if neighbor not in color:
                        color[neighbor] = 1 - color[node]
                        queue.append(neighbor)
                    elif color[neighbor] == color[node]:
                        return False
    return True

print("Graph is Bipartite:", is_bipartite(graph))
```

Graph Connectivity

```python
def is_connected(graph):
    start = next(iter(graph))  
    visited = set()
    queue = deque([start])
    
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            queue.extend(graph[node])
    
    return len(visited) == len(graph)

print("Graph is Connected:", is_connected(graph))
```

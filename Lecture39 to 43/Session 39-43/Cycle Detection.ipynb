### Undirected Graphs: DFS with Parent Tracking

```python
from collections import defaultdict

def has_cycle_undirected(graph):
    visited = set()

    def dfs(node, parent):
        visited.add(node)
        for nei in graph[node]:
            if nei not in visited:
                if dfs(nei, node):
                    return True
            elif nei != parent:      
                return True
        return False

  
    for v in graph:
        if v not in visited:
            if dfs(v, -1):
                return True
    return False

g1 = defaultdict(list, {
    0:[1,2],
    1:[0,2],
    2:[0,1],
    3:[4],
    4:[3]
})

g2 = defaultdict(list, {0:[1],1:[0,2],2:[1,3],3:[2]})

print("Undirected g1 has cycle:", has_cycle_undirected(g1))  
print("Undirected g2 has cycle:", has_cycle_undirected(g2)) 
```

### Directed Graphs: DFS with Recursion Stack

```python
from collections import defaultdict

def has_cycle_directed(graph):
    visiting = set()
    visited = set()   

    def dfs(u):
        if u in visiting:  
            return True
        if u in visited:
            return False

        visiting.add(u)
        for v in graph[u]:
            if dfs(v):
                return True
        visiting.remove(u)
        visited.add(u)
        return False

    for node in graph:
        if node not in visited:
            if dfs(node):
                return True
    return False

gD1 = defaultdict(list, {0:[1], 1:[2], 2:[0]})

gD2 = defaultdict(list, {0:[1], 1:[2], 2:[]})

print("Directed gD1 has cycle:", has_cycle_directed(gD1))  
print("Directed gD2 has cycle:", has_cycle_directed(gD2))  
```

### Cycle Detection Use Cases
Deadlock Detection

```python
wait_for = {
    "P1": ["P2"],
    "P2": ["P3"],
    "P3": ["P1"] 
}
print("Deadlock exists:", has_cycle_directed(wait_for)) 
```

Validating DAGs

```python
prereq = {
    "CourseA": ["CourseB"],
    "CourseB": [],
    "CourseC": ["CourseA"]
}
print("Is DAG (no cycle):",not has_cycle_directed(prereq))
```

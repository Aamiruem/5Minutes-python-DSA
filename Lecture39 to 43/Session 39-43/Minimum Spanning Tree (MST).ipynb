## Prim’s Algorithm

```python
import heapq

def prims_mst(graph, start=0):
    visited = set()
    min_heap = [(0, start)] 
    mst_weight = 0

    while min_heap:
        weight, u = heapq.heappop(min_heap)
        if u in visited:
            continue
        visited.add(u)
        mst_weight += weight

        for v, w in graph[u]:
            if v not in visited:
                heapq.heappush(min_heap, (w, v))

    return mst_weight


graph = {
    0: [(1, 4), (2, 3)],
    1: [(0, 4), (2, 1), (3, 2)],
    2: [(0, 3), (1, 1), (3, 4)],
    3: [(1, 2), (2, 4)]
}
print("Prim's MST weight:", prims_mst(graph))
```

```python
def find_parent(parent, node):
    if parent[node] == node:
        return node
    return find_parent(parent, parent[node])

def union(parent, u, v):
    parent_u = find_parent(parent, u)
    parent_v = find_parent(parent, v)
    parent[parent_v] = parent_u


def get_weight(edge):
    return edge[2]

def kruskal(edges, n):
    edges = sorted(edges, key=get_weight)
    print(edges)
    parent = [i for i in range(n)]
    print(parent)
    mst = []
    total_weight = 0

    for u, v, w in edges:
        if find_parent(parent, u) != find_parent(parent, v):
            union(parent, u, v)
            mst.append((u, v, w))
            total_weight += w

    return mst, total_weight


edges = [
    (0, 1, 4),
    (0, 2, 3),
    (1, 2, 1),
    (1, 3, 2),
    (2, 3, 4)
]

mst, total = kruskal(edges, 4)
print("MST edges:", mst)
print("Total MST weight:", total)
```

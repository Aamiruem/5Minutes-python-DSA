## 1. Constructor & Destructor
* `__init__(self, …)`: Initializes a new object (runs **after** memory is allocated).
* `__del__(self)`: Called **just before** an object is garbage‑collected.

```python
class FileLogger:
    def __init__(self, filename):
        self.filename = filename
        self.file = open(filename, 'w')
        print(f"[INIT] Opened {filename}")

    def log(self, msg):
        self.file.write(msg + '\n')

    def __del__(self):
        print(f"[DEL] Closing {self.filename}")
        self.file.close()

logger = FileLogger('temp.txt')
logger.log('Hello, world!')
del logger
```

## 2. String Representation

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} ({self.age} yrs)"

    def __repr__(self):
        return f"Person(name={self.name!r}, age={self.age})"

p = Person('Shridhar', 30)
print(str(p))
print(repr(p))
```

## 3. Operator Overloading

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, k): 
        return Vector(self.x * k, self.y * k)

    def __eq__(self, other):
        return (self.x, self.y) == (other.x, other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print('v1 + v2  =', v1 + v2)
print('v2 - v1  =', v2 - v1)
print('v1 * 3   =', v1 * 3)
print('v1 == v2?', v1 == v2)
print('v1 == (1,2)?', v1 == Vector(1,2))
```

## 4. Rich Comparisons (`__lt__`, `__gt__`, `__le__`, `__ge__`)

```python
class Box:
    def __init__(self, volume):
        self.volume = volume

    def __lt__(self, other):
        return self.volume < other.volume
    def __le__(self, other):
        return self.volume <= other.volume
    def __gt__(self, other):
        return self.volume > other.volume
    def __ge__(self, other):
        return self.volume >= other.volume
    

b1, b2 = Box(10), Box(20)
print(b1 < b2, b1 > b2)
```

## 5. Callable Objects (`__call__`)
Make an instance behave like a function.

```python
class Adder:
    def __init__(self, increment):
        self.increment = increment

    def __call__(self, value):
        return value + self.increment

five = Adder(5)
print(five(10))
```
